### Implementation of Naive bayes

```python
def prior_pb(y_train,label):
    num = np.sum(y_train==label)
    den = y_train.shape[0]
    return num/den


def cond_pb(X_train,y_train,f,clas,label):
    filtered = X_train[y_train==label]
    num = np.sum(filtered[:,f]==clas)
    den = np.sum(y_train==label)
    return num/den
prior_pb(y_train,1)
```

```python
def predict(x_train,y_train,xtest):
    """Xtest is a single testing point, n features"""
    
    classes = np.unique(y_train)
    pred = []
    n_features = x_train.shape[1]
    post_probs = [] # List of prob for all classes and given a single testing point
    #Compute Posterior for each class
    for label in classes:

            #Post_c = likelihood*prior
            likelihood = 1.0
            for f in range(n_features):
                cond = cond_pb(x_train,y_train,f,xtest[f],label)
                likelihood *= cond

            prior = prior_pb(y_train,label)
            post = likelihood*prior
            post_probs.append(post)
    pred = np.argmax(post_probs)
    return pred
def Naive(X_train,y_train,X_test):
    pred = [] 
    for i in range(len(X_test)):
        pred.append(predict(X_train,y_train,X_test[i]))
    return pred
y_pred = Naive(X_train,y_train,X_test)
```
